```python
# The basic dictionaries

data = {
    "first_name": "gaurav",
    "second_name": "pandey",
    "role": "software_eng",
    "pay": "highest",
    "location": "chennai"
}
```

```python
data
```

```python
data["first_name"]
```

```python
data.get("pay")
```

```python
print(data.get("salary"))
```

```python
type(data.keys())
```

```python
help(data.keys)
```

```python
help(data.values)
```

```python
keys = data.keys()
list_keys = list(data.keys())

print("dict keys: ", keys)
print("list of dict keys: ", list_keys)
```

```python
values = data.values()
list_values = list(data.values())

print("dict values: ", values)
print("list of dict values: ", list_values)
```

```python
data_items = data.items()
list_data_items = list(data.items())

print("dict items: ", data_items)
print("list of dict items: ", list_data_items)
```

```python
data.get("is_admin", False)
```
